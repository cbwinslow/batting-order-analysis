```
# import the Lineup and Simulation classes
from batting_order_analysis.lineup import Lineup
from batting_order_analysis.simulation import Simulation
```

```
# run simulation with a random lineup with 1 game per order
lineup = Lineup()
lineup.set_players()

sim = Simulation(lineup, sims_per_order=1)
sim.run_full_sim()
```

```
# run simulation with a preset lineup using pre-determined outcomes from files in the package contents 
lineup = Lineup()
lineup.set_players(lineup_filename='giants.txt')

sim = Simulation(lineup, pa_outcomes_filename='giants_outcomes.txt')
sim.run_full_sim()
```

```
# run simulation with preset lineup using a list
players = ['PoseyIII,Buster', 'Crawford,Brandon', 'Longoria,Evan', 'Webb,Logan', 'Bryant,Kris', 'LaStella,Tommy', 'Yastrzemski,Mike', 'Flores,Wilmer', 'Ruf,Darin']

lineup = Lineup()
lineup.set_players(lineup_names_arr=players)

sim = Simulation(lineup, sims_per_order=1)
sim.run_full_sim()
```
